#### Google Data Scietist Interview Questions

##### Q1. Given tables for users, logins, and actions, find the number of active users per day (users with at least one login or action).

```python
import numpy as np
import pandas as pd

#Create Example Dataframe
users = {'user_id': [1,2,3],
         'name':['Alice','Bob','Charlie']}
logins = {'user_id':[1, 1, 2,3],
          'login_date':['2024-01-01','2024-01-02','2024-01-02','2024-01-03']}
actions = {'user_id':[1,2,2,3],
           'action_date':['2024-01-01','2024-01-02','2024-01-02','2024-01-03'],
           'action_type':['upload', 'comment', 'like', 'download']}

df_users = pd.DataFrame(users)
df_logins = pd.DataFrame(logins)
df_actions = pd.DataFrame(actions)
```

```python
import sqlite3

# Create an in-memory dataset
conn = sqlite3.connect(":memory:")

# Load into SQLite
df_users.to_sql("users", conn, index=False, if_exists="replace")
df_logins.to_sql("logins", conn, index=False, if_exists="replace")
df_actions.to_sql("actions", conn, index=False, if_exists="replace")

# Run SQL
query = """
SELECT date, COUNT(DISTINCT user_id) AS active_users
FROM(
  SELECT user_id, login_date AS date from logins
  UNION ALL
  SELECT user_id, action_date AS date from actions
)combined
GROUP BY date
"""
result = pd.read_sql_query(query, conn)
print(result)
```
